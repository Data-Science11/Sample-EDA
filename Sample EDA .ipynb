```python
import pandas as pd
import numpy as np
```

```python
dataframe={'Product ID' :['1001', '1002', '1003'],
           'Product Name' :['Mobile PHone', 'USB', 'Earphone'] ,

           'Unit':['20', '30',''],
           'Price' :['20000', '500', ''],
           'Date Time':['20-03-2021  20:23', '03-04-2022  15:25','10-05-2023  14:22']
          }


dataset=pd.DataFrame(dataframe)
dataset.to_csv('data_csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset['Date Time']=pd.to_datetime(dataset['Date Time'], errors='coerce').dt.year
```

```python
dataset['Date Time'].dtype
```

```python
dataset = dataset.replace('', np.nan)
dataset['Price'] = pd.to_numeric(dataset['Price'])
display(dataset)
```

```python
dataset.dtypes
```

```python
dataset['Price'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Unit', data=dataset, hue='Product Name' )
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.barplot(x='Unit', y='Price', data=dataset, hue='Product ID')
plt.show()
```
